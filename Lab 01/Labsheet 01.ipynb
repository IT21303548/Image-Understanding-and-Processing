```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to display images in Jupyter Notebook
def show_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# 1. Read and save image
image = cv2.imread("1.jpg")  # Change "sample.jpg" to your image path
cv2.imwrite("saved_image.jpg", image)  # Saving the image
show_image(image, "Original Image")
```

```python
# 2. Access and modify pixel values
print("Pixel at (100, 100):", image[100, 100])  # Get pixel value
image[100, 100] = [255, 0, 0]  # Modify pixel value
show_image(image, "Modified Pixel Image")
```

```python
# 3. Access image properties
print("Image Shape:", image.shape)  # (Height, Width, Channels)
print("Image Size:", image.size)    # Total pixels
print("Image Datatype:", image.dtype)
```

```python
# 4. Split and Merge Channels
b, g, r = cv2.split(image)
merged_image = cv2.merge([b, g, r])
show_image(merged_image, "Merged Image")
```

```python
# 5. Resize and Rotate Image
resized_image = cv2.resize(image, (200, 200))
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
show_image(resized_image, "Resized Image")
show_image(rotated_image, "Rotated Image")
```

```python
# 6. Draw Shapes
image_shapes = image.copy()
cv2.circle(image_shapes, (100, 100), 50, (0, 255, 0), -1)
cv2.rectangle(image_shapes, (50, 50), (200, 200), (255, 0, 0), 3)
cv2.line(image_shapes, (0, 0), (300, 300), (0, 0, 255), 3)
show_image(image_shapes, "Shapes on Image")
```

```python
# 7. Write Text
image_text = image.copy()
cv2.putText(image_text, "OpenCV", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
show_image(image_text, "Text on Image")
```

```python
# 8. Convert Image to Grayscale, Binary, and Negative
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
negative_image = cv2.bitwise_not(image)
```

```python
show_image(gray_image, "Grayscale Image")
show_image(binary_image, "Binary Image")
show_image(negative_image, "Negative Image")
```
